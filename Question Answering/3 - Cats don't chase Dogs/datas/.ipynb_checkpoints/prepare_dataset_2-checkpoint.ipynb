# dogs chase cats (prepare dataset)

```python
import pickle
from time import time
from discopy import Ty, Word
from discopy.grammar import brute_force
```

```python
s, n = Ty('s'), Ty('n')
```

```python
dogs = Word('dogs', n)
chase = Word('chase', n.r @ s @ n.l)
cats =  Word('cats', n)
_and = Word('and', s.r @ n)
_or = Word('and', s.r @ n)
purr = Word('purr', n.r @ s)

vocab = [dogs, chase, cats, _and, purr]
```

```python
from discopy.grammar import draw

n_sentences = 20
gen = brute_force(*vocab)

dataset_file = open('dataset.csv', 'w')

print("Brute force search for grammatical sentences:\n")

start = time()
for i in range(n_sentences):
    #diagram object for sentence will be serialized in a binary file
    serialization_file = open('serialized_sentences/'+str(i), 'wb')
    #next diagram in bruteforce
    diagram = next(gen)
    sentence = ' '.join(str(w) for w in diagram.boxes if isinstance(w, Word))
    print(diagram)
    draw(diagram, draw_type_labels=False)
    print()
    #update csv
    newline = '\n' if not i == n_sentences-1 else ''
    dataset_file.write('\t'.join([str(i), sentence, newline]))
    #dump diagram object
    dump = pickle.dumps(diagram)
    serialization_file.write(dump)
    serialization_file.close()

dataset_file.close()
print('\nDone !')
```
