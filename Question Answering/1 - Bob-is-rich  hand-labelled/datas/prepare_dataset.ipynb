```python
import pickle
from time import time
from discopy import Ty, Word
from discopy.grammar import brute_force, draw
```

```python
s, n = Ty('s'), Ty('n')
```

```python
Alice = Word('Alice', n)
loves = Word('loves', n.r @ s @ n.l)
Bob =  Word('Bob', n)
who = Word('who', n.r @ n @ s.l @ n)
is_rich = Word('is rich', n.r @ s)

vocab = [Alice, loves, Bob, who, is_rich]
```

```python
dataset_file_name = "dataset.csv"
serialized_sentences_folder = "serialized_sentences"
```

```python
n_sentences = 20
gen = brute_force(*vocab)

#commenting to avoid override
#dataset_file = open(dataset_file_name, 'w')

print("Brute force search for grammatical sentences:\n")

start = time()
for i in range(n_sentences):
    
    #diagram object for sentence will be serialized in a binary file
    serialization_file = open(serialized_sentences_folder+'/'+str(i), 'wb')
    
    #next diagram in bruteforce
    diagram = next(gen)
    sentence = ' '.join(str(w) for w in diagram.boxes if isinstance(w, Word))
    print(diagram)
    draw(diagram, draw_type_labels=False)
    print()
    
    #update csv
    newline = '\n' if not i == n_sentences-1 else ''
    
    #dataset_file.write('\t'.join([str(i), sentence, newline]))
    
    #dump diagram object
    dump = pickle.dumps(diagram)
    serialization_file.write(dump)
    serialization_file.close()

#dataset_file.close()
print('\nDone !')
```

**We label the dataset by hand in a way that the truth value of the sentences are consistent**

```python
sentences2truth_values = {}
with open(dataset_file_name) as fp:
    for line in fp:
        line = line.rstrip('\n')
        sent_id, sent, truth_val = line.split('\t')
        sentences2truth_values[sent] = int(truth_val)

print("True sentences:\n{}\n".format('\n'.join(sentence
    for sentence, truth_value in sentences2truth_values.items() if truth_value)))
print("False sentences:\n{}".format('\n'.join(sentence
    for sentence, truth_value in sentences2truth_values.items() if not truth_value)))
```
