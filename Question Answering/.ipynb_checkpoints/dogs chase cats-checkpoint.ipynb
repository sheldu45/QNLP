# dogs chase cats

This experiment is inspired by bob-is-rich experiment at QNLP 2019.

```python
import numpy as np
#jax: library from Google which allows to compute gradient of any python function (in our case the function that evaluates the Qu Circuit)
from jax import jit
import jax.numpy as np
from time import time
import pickle
from sklearn.model_selection import train_test_split
from random import random, uniform, seed; 
from discopy import Tensor
from discopy import Ty, Word, CircuitFunctor, qubit
from discopy.quantum import CX, H, Ket, Rx, Id, sqrt
from discopy.quantum.circuit import Functor
from discopy.rigid import Spider 
```

**1) First, we define the atomic pregroup types (typically ‘s’ and ‘n’).**

```python
s, n = Ty('s'), Ty('n')
```

**2) We assign a number of qubits for each pregroup type.**

```python
ob = {s: 1, n: 1}
```

**3) Then we construct the finite vocabulary. For each word we assign a pregroup type.**

```python
dogs = Word('dogs', n)
chase = Word('chase', n.r @ s @ n.l)
cats =  Word('cats', n)
_and = Word('and', s.r @ n)
purr = Word('purr', n.r @ s)

vocab = [dogs, chase, cats, _and, purr]
```

**4) Every pregroup type will be mapped to a Quantum Circuit. We explicitely build the Quantum Circuits.**

Circuit for the GHZ state, which modelizes relative pronouns

```python
#"Special Quantum State", there is a lot of litterature (?) on how relative pronouns (i.e. who, that...) can be modelized nicely using GHZ
'''GHZ = Functor(ob=lambda o: o, ar=lambda a: a)(Spider(0, 3, s, phase=0.5))

GHZ.draw(aspect='auto', draw_type_labels=False)
print(GHZ.eval())'''
```

Circuit for the intransitive ansatz state:

```python
#is-rich
def intransitive_ansatz(phase):
    return sqrt(2) @ Ket(0, 0) >> H @ Rx(phase) >> CX

iv = intransitive_ansatz(0.5)
iv.draw(aspect='auto', draw_type_labels=False)
print(iv.eval())
```

Circuit for the transitive ansatz state:

```python
#will input 2 qbits, output  a scalar (result of measurement)
'''
def transitive_ansatz(phase):
    return Functor(ob=lambda o: o, ar=lambda a: a)(Spider(0, 3, s, phase=phase))
'''
def transitive_ansatz():
    return Functor(ob=lambda o: o, ar=lambda a: a)(Spider(0, 3, s))

tv = transitive_ansatz()
tv.draw(aspect='auto', draw_type_labels=False)
print(tv.eval())
```

**5) We explicit the mapping for each entry in the vocabulary to their corresponding Quantum Circuits.**

```python
ar = lambda params: {
    dogs: Ket(0),
    #chase: transitive_ansatz(params['chase']),
    chase: transitive_ansatz(),
    cats: Ket(1),
    _and: intransitive_ansatz(params['_and']),
    purr: intransitive_ansatz(params['purr'])}
```

**6) Because the circuit is parameterized, we define a function which takes a set of parameters as inputs and outputs a functor from the grammar category to the category of quantum circuits**

```python
#F is a function which takes parameters and constructs a Functor
#F = lambda params: CircuitFunctor(ob, ar({'chase': params[0], '_and': params[1], 'purr': params[2]}))
F = lambda params: CircuitFunctor(ob, ar({'_and': params[1], 'purr': params[2]}))

'''
serialized_diagram = open('datas/serialized_sentences/11', 'rb')
diagram = pickle.loads(serialized_diagram.read())
print(diagram)

print("Circuit 'dogs chase cats.' :")
#We apply the functor to 'cats chase dogs and purr.' 
F(params0)(diagram).draw(figsize=(12, 8), aspect='auto', draw_type_labels=False, nodesize=1)'''
```

**7) Another functor is created to evaluate the circuit**

```python
#We evaluate the circuit through applying another functor: 
# a qubit (2 dim) -> vector space, a gate -> Matrix (ie. Rx)

evaluate = lambda F, sentence: F(sentences2diagram[sentence]).measure()
```

We download the csv

```python
sentences = []
sentences2truth_values = {}
sentences2diagram = {}

with open('datas/dataset.csv') as fp:
    for line in fp:
        line.strip("\n")
        sent_id, sentence, truth_value = line.split('\t')
        serialized_diagram = open('datas/serialized_sentences/'+sent_id, 'rb')
        diagram = pickle.loads(serialized_diagram.read())
        sentences.append(sentence)
        sentences2truth_values[sentence] = int(truth_value)
        sentences2diagram[sentence] = diagram
        #for visualization
        print(sentence)
        F(np.array([1., 1.]))(diagram).draw(figsize=(12, 8), aspect='auto', draw_type_labels=False, nodesize=1)
        print()
        
```

Sentences are arbitrarily (with assignement of params0) set to True or False

We split in train/test

```python
sentence_train, sentence_test = train_test_split(sentences, test_size=0.3, random_state=42)

print("Training set:\n{}\n".format('\n'.join(sentence_train)))
print("Testing set:\n{}".format('\n'.join(map(str, sentence_test))))
```

```python
#we now evaluate every sentence of corpus using the constructed functor
corpus = sentences2truth_values#{sentence: evaluate(F(params0), sentence) for sentence in sentences}

epsilon = 1e-2

print("True sentences:\n{}\n".format('\n'.join(sentence
    for sentence, probability in corpus.items() if probability > 1 - epsilon)))
print("False sentences:\n{}".format('\n'.join(sentence
    for sentence, probability in corpus.items() if probability < epsilon)))
```

We recover parameters that maximizes true/false prediction with gradient descent

```python
seed(0)
Tensor.np = np  # This ensures we can differentiate Circuit.eval with respect to phases.

@jit
def training_loss(params):
    return np.mean(np.array([
        (corpus[sentence] - evaluate(F(params), sentence)) ** 2
        for sentence in sentence_train]))

@jit
def testing_loss(params):
    return np.mean(np.array([
        (corpus[sentence] - evaluate(F(params), sentence)) ** 2
        for sentence in sentence_test]))

#apply loss function on every circuit
'''start = time()
print("Training loss: {}".format(training_loss(params0)))
print("Testing loss: {}".format(testing_loss(params0)))
print("\n{:.0f} seconds to compile the loss.".format(time() - start))'''
```

```python
from jax import grad

step_size = 1e-2

@jit
def update(params):
    return params - step_size * grad(training_loss)(params)

'''start = time()
p_loves, p_is_rich = update(params0)
print("params['loves'] = {}".format(p_loves))
print("params['is rich'] = {}".format(p_is_rich))
print("\n{:.0f} seconds to compile the update.".format(time() - start))'''
```

```python
#we forget the true parameters (used to generate the corpus) and restart learning them from scratch (for love and is rich)
params = np.array([uniform(-1, 1), uniform(-1, 1), uniform(-1, 1)])

print("Random initialisation: {}".format(params))
print("Initial training loss: {}".format(training_loss(params)))
print("Initial testing loss: {}".format(testing_loss(params)))

print("\ndogs chase cats and pur?")
print(evaluate(F(params), 'dogs chase dogs.') )
```

```python
def funct_evaluate_sentence(sent):
    tested_sentence = sent.rstrip('.\n')
    output_qbit = evaluate(F(params), tested_sentence+'.')
    prediction = 0 if output_qbit[0] > output_qbit[1] else 1
    print(tested_sentence+'?', end='\t')
    print("Yes."
      if prediction
      else "No.", end='\t')
    print("Correct" 
      if sentences2truth_values[tested_sentence+'.'] == prediction
      else 'Incorrect'
    )
```

```python
#"vanilla" gradient descent
epochs, iterations = 3, 100

for i in range(epochs):
    start_epoch = time()

    for j in range(iterations):
        start_iteration = time()
        params = update(params)

        '''if i == 0 and j < 3:
            print("Update {} ({:.0f} microseconds)".format(j, 1e6 * (time() - start_iteration)))
            print("params['_chase'] = {:.3f}".format(params[0]))
            print("params['_and'] = {:.3f}".format(params[1]))
            print("params['purr'] = {:.3f}".format(params[2]))
            print("Training loss: {}".format(training_loss(params)))
            print("Testing loss: {}\n".format(testing_loss(params)))'''

    print("Epoch {} ({:.0f} milliseconds)".format(i, 1e3 * (time() - start_epoch)))
    print("params['_and'] = {:.3f}".format(params[0]))
    print("params['purr'] = {:.3f}".format(params[1]))
    '''print("params['chase'] = {:.3f}".format(params[0]))
    print("params['_and'] = {:.3f}".format(params[1]))
    print("params['purr'] = {:.3f}".format(params[2]))'''
    print("Training loss: {}".format(training_loss(params)))
    print("Testing loss: {}\n".format(testing_loss(params)))
```

```python
i = 1
for sent, val in sentences2truth_values.items():
    print(i, end='\t')
    funct_evaluate_sentence(sent)
    i+=1
```

```python
from jax import vmap
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
start = time()

x, y = np.meshgrid(np.arange(-1.0, 1.0, 1e-2), np.arange(-1.0, 1.0, 1e-2))
xy = np.dstack((x.flatten(), y.flatten()))[0]
z = vmap(training_loss)(xy).reshape(x.shape)

print("{:.3f} seconds to compute the landscape.".format(time() - start))

#z axis: loss function 
fig = plt.figure(figsize=(10, 10))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm)
plt.xlabel("params['loves']")
plt.ylabel("params['is rich']")
plt.title("functorial landscape for '_and' and 'purr'")
plt.show()
```
